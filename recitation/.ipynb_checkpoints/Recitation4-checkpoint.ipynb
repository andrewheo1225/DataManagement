```python
# Find the most popular genres by counting
# the number of movies created for each genre the print them
def find_popular_genres():
  genres_dict = {}
  for line in open("/content/ml-latest-small/movies.csv"):
    l = line.rstrip('\n').split(',')
    genres = l[2]
    for g in genres.split('|'):
      if g in genres_dict:
        genres_dict[g] += 1
      else:
        genres_dict[g] = 1
  print(genres_dict.items())
  sorted_genres = sorted(genres_dict, key = lambda genre: genres_dict[genre], reverse=True)
  print(sorted_genres[0:10])
```

```python
# find the average rating for each movie, then 
# print the top 5 movies with the highest ratings
def find_popular_movies():
  count = 0
  ratings_dict = {}
  avg_ratings = {}
  for line in open("/content/ml-latest-small/ratings.csv"):
    l = line.rstrip('\n').split(',')
    id = l[1]
    rating = l[2]

    if id in ratings_dict:
        ratings_dict[id].append(rating)
    else:
      ratings_dict[id] = [rating]
  print(ratings_dict)
  ratings_dict.pop('movieId')
  for k in ratings_dict:
    my_list = [float(r) for r in ratings_dict[k]]
    avg_ratings[k] = sum(my_list)/len(my_list)
  print(avg_ratings)
```
