###Functions

A function is a block of code which only runs when it is called. You can pass data, known as parameters, into a function.

```python
# function with single parameter
def print_name(name):
  print("Hello, My name is: ", name)

print_name("Mike")
print_name("Jake")
print_name("Jim")
```

```python
# function with default parameters
def my_function(country = "Norway"):
  print("I am from " + country)

my_function("Sweden")
my_function("India")
my_function()
my_function("Brazil")
```

```python
# function with both default and other parameters
def func_multiply(x,y=5):
    return x*y

print(func_multiply(3))
```

```python
# return functions 
def operate(a, b):
    sum = a + b
    diff = a - b
    mul = a * b
    div = a / b
    return sum, diff, mul, div

n1 = 5
n2 = 10
sum, diff, mul, div = operate(n1, n2)
print(
    f"The sum is {sum}\n"
    f"The difference is {diff}\n"
    f"The multiplication gives {mul}\n"
    f"The division gives {div}\n"
)
```

### Loops
Python has two primitive loop commands:

* while loops
* for loops

```python
# while loop example
i = 1
while i < 6:
  print(i)
  i += 1

for i in range(1,6):
  print(i)
```

```python
# exit loop when i is 3
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

```python
# With the else statement we can run a block of code once when the condition no longer is true:
i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("i is no longer less than 6")
```

```python
# for loop using range function
for x in range(3,6):
  print(x)
```

```python
# for loop from 2 to 5
for x in range(2, 6):
  print(x)
```

```python
# increment the sequence with 3 (default is 1):

for x in range(30, 3, -3):
  print(x)
```

```python
# iterating over string
# count number of vowels and consonants in a string
def countVowelsAndConsonants(str):
    v_count=0
    c_count=0
    for ch in str:  # cycles through each character in the string, in is membership operator
        if ch.lower() in 'aeiou': # checks for membership in vowel list
            v_count += 1
        else:
            c_count += 1
    return v_count, c_count

vc, cc = countVowelsAndConsonants("And the quick brown fox jumped over the lazy dog")
print(f'#vowels={vc}, #consonants={cc}')
```

### Error Handling

When an error occurs, or exception as we call it, Python will normally stop and generate an error message.

```python
# Throw error when input is not an integer
sum = 0
while True:
    try:
        val = int(input('Enter next integer, 0 to stop: '))
        if val == 0:
            break
        sum += val
    except ValueError:
        print("Input must be an integer")
print(sum)
```
